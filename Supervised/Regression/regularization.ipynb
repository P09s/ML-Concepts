```python
import numpy as np

X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])

def gradient_descent_reg(X, y, alpha, iterations, lambda_):
    m, n = X.shape
    
    # initialize
    w = np.zeros(n)
    b = 0
    
    for i in range(iterations):
        
        # predictions
        f_wb = X @ w + b
        
        # error
        error = f_wb - y
        
        # cost (with regularization)
        cost = (1/(2*m)) * np.sum(error**2) + (lambda_/(2*m)) * np.sum(w**2)
        
        # gradients
        dj_dw = (1/m) * (X.T @ error) + (lambda_/m) * w
        dj_db = (1/m) * np.sum(error)
        
        # update
        w -= alpha * dj_dw
        b -= alpha * dj_db
        
        # print progress
        if i % 100 == 0:
            print(f"Iteration {i}: Cost = {cost}")
    
    return w, b
```

```python
w, b = gradient_descent_reg(X_train, y_train, alpha = 0.01, iterations=2000, lambda_=0.2)

print(f"w: {w}, b: {b}")
```
