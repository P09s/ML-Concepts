```python
'''
here we'll use dot product based on the vectorization concept to 
implement our gradient descent and compute values from a regression model
in multiple features
'''
```

```python
import numpy as np
```

```python
x_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y_train = np.array([460, 232, 178])
```

```python
def compute_cost(X,y,w,b):
    m = X.shape[0]
    cost = 0.0

    for i in range(m):
        f_wb = np.dot(X[i], w) + b
        cost = cost + (f_wb - y[i])**2

    total_cost = cost/(2*m)
    return total_cost
```

```python
def compute_gradient(X, y, w, b):
    m = X.shape[0]
    dj_dw = np.zeros_like(w)   # ✅ MUST be vector
    dj_db = 0.0

    for i in range(m):
        f_wb = np.dot(X[i], w) + b
        error = f_wb - y[i]

        dj_dw += error * X[i]
        dj_db += error

    dj_dw = dj_dw / m
    dj_db = dj_db / m

    return dj_dw, dj_db
```

```python
def gradient_descent(X, y, w_in, b_in, alpha, num_iters, compute_gradient):
    w = w_in
    b = b_in

    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient(X, y, w, b)
        w = w - (alpha * dj_dw)
        b = b - (alpha * dj_db)

    return w, b
```

```python
w_init = np.zeros(x_train.shape[1])
b_init = 0.0

iterations = 1000
alpha = 1e-7

w_final, b_final = gradient_descent(x_train, y_train, w_init, b_init, alpha, iterations, compute_gradient)

print(f"(w,b) founded by Gradient Descent are w = {w_final}, b = {b_final}")
```
