```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train: {x_train}")
print(f"y_train: {y_train}")
```

```python
print(f"train shape: {x_train.shape}")
m = x_train.shape[0]
print(f"no. of training examples: {m}")
```

```python
i = 0

x_i = x_train[i]
y_i = y_train[i]

print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```
