```python
import numpy as np
```

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480, 430, 630, 730])
```

```python
def compute_cost(x, y, w, b):
    m = x.shape[0]
    cost = 0

    for i in range(m):
        f_wb = w*x[i] + b
        cost = cost + (f_wb - y[i])**2
    total_cost = cost/(2*m)
    return total_cost 
```

```python
def compute_gradient(x, y, w, b):
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        f_wb = w*x[i] + b

        dj_dw_i = (f_wb - y[i])*x[i]
        dj_db_i = f_wb - y[i]

        dj_dw = dj_dw + dj_dw_i
        dj_db = dj_db + dj_db_i
    
    dj_dw = dj_dw/m
    dj_db = dj_db/m

    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters,compute_gradient):
    w = w_in
    b = b_in

    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient(x, y, w, b)

        w = w - alpha * dj_dw
        b = b - alpha * dj_db
    return w, b
```

```python
w_init = 0
b_init = 0

iterations = 1000
alpha = 1.0e-2

w_final, b_final = gradient_descent(x_train, y_train, w_init, b_init, alpha, iterations, compute_gradient)

print(f"(w,b) founded by Gradient Descent are w = {w_final: .2f}, b = {b_final: .2f}")
```
