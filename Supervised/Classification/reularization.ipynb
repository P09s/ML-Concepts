```python
import numpy as np
```

```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])
```

```python
def gradient_descent(X, y, alpha, iterations, lambda_):
    m,n = X.shape
    w = np.zeros(n)
    b = 0

    for i in range(iterations):
        z = X @ w + b
        f_wb = 1/(1+np.exp(-z))
        error = f_wb - y

        cost = (-1/m)*np.sum(y*np.log(f_wb) + (1-y)*np.log(1-f_wb)) \
               + (lambda_/(2*m)) * np.sum(w**2)

        dj_dw = (1/m)*(X.T@error) + (lambda_/m)*w
        dj_db = (1/m)*np.sum(error)

        w -= alpha * dj_dw
        b -= alpha * dj_db

        if i % 100 == 0:
            print(f"For {i} iterations: Cost = {cost}")

    return w,b
```

```python
w, b = gradient_descent(X_train, y_train, alpha = 0.01, iterations=2000, lambda_=0.2)

print(f"w: {w}, b: {b}")
```
