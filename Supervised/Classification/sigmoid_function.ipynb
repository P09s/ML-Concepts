```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def sigmoid_function(z):
    return 1/(1 + np.exp(-z))
```

```python
z_values = np.array([-10, -5, -1, 0, 1, 5, 10])
result_y = sigmoid_function(z_values)

print(result_y)
```

```python
# Or we can print in a map view too !!

for z, r in zip(z_values, result_y):
    print(f"{z} -> {r:.2f}")
```
