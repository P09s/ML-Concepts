```python
import numpy as np
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
def compute_cost_logistic(X, y, w, b):
    m = X.shape[0]
    cost = 0.0

    for i in range(m):
        z = np.dot(X[i], w) + b
        f = sigmoid(z)

        cost += -y[i] * np.log(f) - (1 - y[i])*np.log(1-f)
    return cost/m
```

```python
X = np.array([
    [0.5, 1.5],
    [1, 1],
    [1.5, 0.5],
    [3, 0.5],
    [2, 2],
    [1, 2.5]
])

y = np.array([0, 0, 0, 1, 1, 1])
```

```python
# Case 1 (better)
w1 = np.array([1, 1])
b1 = -3

# Case 2 (worse)
w2 = np.array([1, 1])
b2 = -4

cost1 = compute_cost_logistic(X, y, w1, b1)
cost2 = compute_cost_logistic(X, y, w2, b2)

print("Cost for b = -3:", cost1)
print("Cost for b = -4:", cost2)
```
