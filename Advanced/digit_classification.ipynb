```python
import numpy as np 
import tensorflow as tf 
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

```python
data = load_digits()

X = data.data
y = data.target

X = X/16.0

X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=0.2, random_state=42
)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
model = Sequential([
    tf.keras.Input(shape=(64,)),
    Dense(units=64, activation='relu'),
    Dense(units=32, activation='relu'),
    Dense(units=10, activation='linear')
])
```

```python
model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer = 'adam',
    metrics = ['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs=30, validation_split=0.1)
```

```python
loss, acc = model.evaluate(X_test, y_test)
print("\n Test Accuracy:", acc)
```

```python
sample = X_test[0].reshape(1, 64)

logits = model.predict(sample)
probs = tf.nn.softmax(logits).numpy()
pred_class = np.argmax(probs)

print("\nPredicted:", pred_class)
print("Actual:", y_test[0])
print("Probabilities:", probs)
```
