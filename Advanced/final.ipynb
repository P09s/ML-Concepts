```python
# =========================
# 1. IMPORTS
# =========================
import numpy as np


# =========================
# 2. DATASET
# =========================
X_train = np.array([
    [1,1,1],
    [1,0,1],
    [1,0,0],
    [1,0,0],
    [1,1,1],
    [0,1,1],
    [0,0,0],
    [1,0,1],
    [0,1,0],
    [1,0,0]
])

y_train = np.array([1,1,0,0,1,0,0,1,1,0])


# =========================
# 3. ENTROPY
# =========================
def compute_entropy(y):
    if len(y) == 0:
        return 0

    p1 = np.sum(y == 1) / len(y)

    if p1 == 0 or p1 == 1:
        return 0

    return -p1 * np.log2(p1) - (1 - p1) * np.log2(1 - p1)


# =========================
# 4. SPLIT DATASET
# =========================
def split_dataset(X, node_indices, feature):
    left_indices = []
    right_indices = []

    for i in node_indices:
        if X[i][feature] == 1:
            left_indices.append(i)
        else:
            right_indices.append(i)

    return left_indices, right_indices


# =========================
# 5. INFORMATION GAIN
# =========================
def compute_information_gain(X, y, node_indices, feature):

    left_indices, right_indices = split_dataset(X, node_indices, feature)

    y_node = y[node_indices]
    y_left = y[left_indices]
    y_right = y[right_indices]

    # Entropy
    H_node = compute_entropy(y_node)
    H_left = compute_entropy(y_left)
    H_right = compute_entropy(y_right)

    # Weights
    w_left = len(y_left) / len(y_node)
    w_right = len(y_right) / len(y_node)

    # Information Gain
    IG = H_node - (w_left * H_left + w_right * H_right)

    return IG


# =========================
# 6. BEST SPLIT
# =========================
def get_best_split(X, y, node_indices):

    num_features = X.shape[1]
    best_feature = -1
    max_IG = 0

    for feature in range(num_features):
        IG = compute_information_gain(X, y, node_indices, feature)

        if IG > max_IG:
            max_IG = IG
            best_feature = feature

    return best_feature


# =========================
# 7. BUILD TREE (RECURSIVE)
# =========================
def build_tree_recursive(X, y, node_indices, depth, max_depth):

    # Stop condition
    if depth == max_depth:
        print("  " * depth + f"Leaf node → indices {node_indices}")
        return

    # Get best feature
    best_feature = get_best_split(X, y, node_indices)

    print("  " * depth + f"Depth {depth}: Split on feature {best_feature}")

    # Split
    left_indices, right_indices = split_dataset(X, node_indices, best_feature)

    # Recurse
    build_tree_recursive(X, y, left_indices, depth + 1, max_depth)
    build_tree_recursive(X, y, right_indices, depth + 1, max_depth)


# =========================
# 8. RUN EVERYTHING
# =========================
root_indices = list(range(len(X_train)))

print("=== Building Decision Tree ===\n")
build_tree_recursive(X_train, y_train, root_indices, depth=0, max_depth=2)
```
