```python
import numpy as np 
import tensorflow as tf
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense
```

```python
# Dummy coffee dataset (since lab utils not available)
X = np.array([
    [200, 12],
    [250, 10],
    [180, 15],
    [300, 5],
    [220, 13]
])

Y = np.array([[1], [1], [1], [0], [1]])  # 1 = good, 0 = bad
```

```python
# Model

model = Sequential([
    Dense(units=3,activation='relu'),
    Dense(units=1,activation='sigmoid')
])
```

```python
# Compile

model.compile(
    loss='binary_crossentropy',
    optimizer='adam'
)
```

```python
# Train

model.fit(X, Y, epochs=100)
```

```python
# Predict
test = np.array([[210, 11]])
pred = model.predict(test)

print(f"Prediction = {pred}")

if pred >= 0.5:
    print('We will get a Good Coffee')
else:
    print('We will get a Bad Coffee')
```
