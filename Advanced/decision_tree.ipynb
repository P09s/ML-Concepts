```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from utils import *
```

```python
def entropy(p):
    if p == 0 or p == 1:
        return 0
    return -p*np.log2(p) - (1-p)*np.log2(1-p)

p_vals = np.linspace(0, 1, 100)
H_vals = [entropy(p) for p in p_vals]

plt.plot(p_vals, H_vals)
plt.xlabel("Probability (p)")
plt.ylabel("Entropy H(p)")
plt.title("Entropy vs Probability")
plt.show()
```

```python
def information_gain(parent, left, right):
    H_parent = entropy(parent)

    w_left = len(left)/(len(left) + len(right))
    w_right = len(right)/(len(left) + len(right))

    H_left = entropy(np.mean(left))
    H_right = entropy(np.mean(right))

    IG = H_parent - (w_left*H_left + w_right*H_right)
    return IG
```

```python
parent = 0.5
left = [1,1,1,0]
right = [0,0,0]

print(information_gain(parent, left, right))
```
