```python
import numpy as np
```

```python
def gradient_descent(X, y, alpha=0.01, iterations=5000):
    m, n = X.shape
    w = np.zeros(n)
    b = 0

    for i in range(iterations):
        f_wb = X @ w - b
        error = f_wb - y

        dj_dw = (1/m)*(X.T @ error)
        dj_db = (1/m)*np.sum(error)

        w =- alpha*dj_dw
        b =- alpha*dj_db

    return w, b
```

```python
# DATASET (Manual)

x = np.arange(0, 20, 1)
y = x**2
```

```python
#feature engineering

X1 = np.c_[x, x**2, x**3]
```

```python
#normalization

mu = np.mean(X1, axis=0)
sigma = np.std(X1, axis=0)
X1 = (X1 - mu)/sigma

w, b = gradient_descent(X1, y, alpha=0.1)
pred = X1 @ w + b

print(f"w= {w}, b = {b:.2f}")
print(pred)
```
