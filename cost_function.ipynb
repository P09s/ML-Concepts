```python
import numpy as np
```

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480, 430, 630, 730])
```

```python
def compute_cost(x, y, w, b):
    m = x.shape[0]
    cost_sum = 0

    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i])**2
        cost_sum = cost_sum + cost
        total_cost = cost_sum / (2*m)
    return total_cost
```

```python
compute_cost(x_train, y_train, 100, 100)
```
