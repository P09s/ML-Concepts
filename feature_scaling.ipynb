```python
import numpy as np
```

```python
X = np.array([
    [952, 2, 1, 65],
    [1244, 3, 2, 64],
    [1947, 3, 2, 17],
    [1720, 3, 2, 42],
    [2000, 4, 2, 20],
    [850, 2, 1, 70],
    [1100, 3, 1, 50],
    [1600, 3, 2, 30],
    [2400, 4, 3, 10],
    [3000, 5, 3, 5]
])

y = np.array([
    271.5, 232, 509.8, 340, 600,
    250, 310, 400, 720, 900
])
```

```python
# Normalization

def normalize(X):
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    X_norm = (X - mu)/sigma
    return X_norm, mu, sigma
```

```python
def gradient_descent(X, y, alpha = 0.1, iterations = 200):
    m, n = X.shape
    w = np.zeros(n)
    b = 0

    cost_history = []

    for i in range(iterations):
        f_wb = X @ w - b
        error = f_wb - y

        dj_dw = (1/m) * (X.T @ error)
        dj_db = (1/m) * np.sum(error)

        w =- alpha*dj_dw
        b =- alpha*dj_db

        cost = (1/(2*m)) * np.sum(error**2)
        cost_history.append(cost)

        if(i % 50 == 0):
            print(f"Iter {i}: Cost {cost:.2f}")
        
        return w, b, cost_history

```

```python
# 📌 Transpose (X.T)
# X shape: (m, n) → m examples, n features
# X.T shape: (n, m) → n features, m examples
# 👉 It swaps rows and columns

# 📌 Why we use X.T @ error
# error has shape (m,) (one error per example)
# X.T @ error → shape (n,)
# 👉 Gives one gradient per feature (w1, w2, ..., wn)

# 📌 Intuition
# X → grouped by examples
# X.T → grouped by features
# 👉 We need gradients per feature → use X.T

# 📌 One-line summary
# X.T flips the matrix so we can compute gradients 
# for each feature using all training examples at once.
```

```python
X_norm, mu, sigma = normalize(X)
w, b, _ = gradient_descent(X_norm, y)

x_test = np.array([1200, 3, 1, 40])

X_norm = (x_test - mu)/sigma
pred_correct = X_norm @ w + b

print(f"wait w = {w} and bias b = {b:.2f}")
print(f"predicted value is {pred_correct:.2f}")
```
